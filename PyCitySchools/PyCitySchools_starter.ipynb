```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# calculate number of schools
total_schools = school_data_complete.school_name.nunique()
total_schools
```

```python
#calculate number of students
total_students = len(school_data_complete.index)
total_students
```

```python
#listing the school budgets
unique = school_data_complete["budget"].unique()
unique
```

```python
#total budget of the schools
total_budget = unique.sum()
total_budget
```

```python
#average math score
avr_math = school_data_complete["math_score"].mean()
avr_math
```

```python
#average reading score
avr_reading = school_data_complete["reading_score"].mean()
avr_reading
```

```python
# Create a new column that finds the overall average score
col = school_data_complete.loc[: , "reading_score":"math_score"]
school_data_complete['overall_passing_rate'] = col.mean(axis=1)
school_data_complete
```

```python
#calculating math passing
school_data_complete['math_passing'] = np.where(school_data_complete['math_score']>=70, 'yes', 'no')
school_data_complete
```

```python
#math passing percentage
math_passing = school_data_complete['math_passing'].value_counts(normalize=True) * 100
math_passing = math_passing[0]
math_passing
```

```python
#calculating reading passing
school_data_complete['reading_passing'] = np.where(school_data_complete['reading_score']>=70, 'yes', 'no')
school_data_complete
```

```python
# reading passing percentage
reading_passing = school_data_complete['reading_passing'].value_counts(normalize=True) * 100
reading_passing = reading_passing[0]
reading_passing
```

```python
#calculating overall passing
school_data_complete['overall_passing'] = np.where(school_data_complete['overall_passing_rate']>=70, 'yes', 'no')
school_data_complete
```

```python
# overall passing percentage
overall_passing = school_data_complete['overall_passing'].value_counts(normalize=True) * 100
overall_passing = overall_passing[0]
overall_passing
```

```python

# Place all of the data found into a summary DataFrame

summary_table = pd.DataFrame({"total schools": [total_schools],
                                      "total students": [total_students],
                                      "total budget": [total_budget],
                                      "avr math score": [avr_math],
                                      "avr reading score":[avr_reading],
                                      "passing math": [math_passing],
                                      "passing reading": [reading_passing],
                                      "overall passing": [overall_passing]})
summary_table
```
